```
import os
list_of_files = os.listdir()
```

```
from tqdm import tqdm
import json
```

```
list_of_abstracts = []
list_of_body = []
counter = 0
for i in tqdm(list_of_files):
    with open(i, "rb") as f:
        s = json.load(f)
    if len(s['abstract']) != 0:
        list_of_abstracts.append(s['abstract'][0]['text'])
        entire_body = ""
        for j in s['body_text']:
            entire_body += j['text'] + " \n\n "  
        list_of_body.append(entire_body)
```

```
import re
reference_1 = r"\[[^d]*\]"
figures = r"figure \d[a-zA-Z]"
reference_2 = r"\(\d+\)"
new_list_of_body = []
for i in tqdm(list_of_body):
    a = re.sub(reference_1, '', i)
    b = re.sub(figures, '', a)
    c = re.sub(reference_2, '', b)
    new_list_of_body.append(c)
```

```
import pandas as pd
main_df = pd.DataFrame()
main_df["Abstract"] = list_of_abstracts
main_df["Body"] = new_list_of_body
```
